```python
import requests

def download_pdfs(pdf_urls):
  """Downloads PDFs from a list of URLs and saves them locally."""
  for url in pdf_urls:
    response = requests.get(url)
    filename = url.split('/')[-1]  # Extract filename from URL
    with open(filename, 'wb') as f:
      f.write(response.content)

pdf_urls = pdf_urls_string.splitlines()  # Split string into list
download_pdfs(pdf_urls)
```

```python
import os
from PyPDF2 import PdfReader
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define the source directory containing the PDF files
source_directory = '/content/'

# Define the destination directory in Google Drive
destination_directory = '/content/drive/MyDrive/TXT_Files/'

# Create the destination directory if it doesn't exist
os.makedirs(destination_directory, exist_ok=True)

# Iterate over all files in the source directory
for filename in os.listdir(source_directory):
    if filename.endswith('.pdf'):  # Check if the file is a PDF file
        # Open the PDF file
        with open(os.path.join(source_directory, filename), 'rb') as pdf_file:
            pdf_reader = PdfReader(pdf_file)
            # Extract text from each page of the PDF file
            text = ''
            for page in pdf_reader.pages:
                text += page.extract_text()

        # Define the destination TXT file path
        txt_filename = os.path.splitext(filename)[0] + '.txt'
        txt_file_path = os.path.join(destination_directory, txt_filename)

        # Write the extracted text to the TXT file
        with open(txt_file_path, 'w') as txt_file:
            txt_file.write(text)

        print(f"PDF file '{filename}' converted and saved as TXT file '{txt_filename}'.")

print("All PDF files converted and saved as TXT files to Google Drive.")
```

```python
from bs4 import BeautifulSoup
import urllib.request

# URLs of the pages to scrape
urls = [
    "https://extension.colostate.edu/staff-directory/?cn-s=&cn-cat=113",
    "https://extension.colostate.edu/staff-directory/pg/2/?cn-cat=113&cn-s",
    "https://extension.colostate.edu/staff-directory/pg/3/?cn-cat=113&cn-s"
]

# Loop through each URL
for url in urls:
    # Send a GET request to the URL
    req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    response = urllib.request.urlopen(req)

    # Check if the request was successful
    if response.status == 200:
        # Extract the HTML content
        html_content = response.read().decode('utf-8')

        # Parse the HTML content
        soup = BeautifulSoup(html_content, "html.parser")

        # Find all div elements with class "cn-entry"
        divs = soup.find_all("div", class_="cn-entry")

        # Iterate over the div elements and extract the address components
        for div in divs:
            # Extract all p elements within the current div
            paragraphs = div.find_all("p")

            # Initialize variables to store address components
            address = ""
            for paragraph in paragraphs:
                # Concatenate the text of p elements to form the address
                address += paragraph.text.strip() + " "

            print("Address:", address.strip())
            print()  # Add a new line between addresses

    else:
        print("Failed to retrieve data from the URL:", url)
```
